```r
library(mongolite)
library(tidyverse)
library(quanteda)
library(lubridate)
```

```r
collection <- "politicians_sample" # or "organizations" "selectors"
username <- "btw17_public_user"
password <- "btw17_public"
ip <- "10.6.13.55"
db = "twitter_public"
     
```

```r
m <- mongo(collection = collection, url =
             paste0('mongodb://',username,':',password,'@',ip,'/',db))
```

```r
df <- m$find(
  query = '{
 }',
  fields = '{"_id": 1, "user.id": 1, "user.name": 1, "user.screenName": 1, "createdAt": 1, "text": 1, "source": 1}'
)

     
```

```r
a = df %>%
    mutate(date = round_date(createdAt, "minute")) %>%
    group_by(date) %>%
    summarise(n = n()) %>%
    ggplot(data = ., aes(x  = date, y = n)) +
           geom_line()+
    theme_bw()
```

```r
print(a)
```

```r
## Accounts with most retweets over time:

df$retweet <- ifelse(grepl("^RT ", df$text), 1, 0) 
#table(df$retweet)

head(df$text)
```

```r
retweet_df <- df %>%
    filter(retweet == 1) %>%
    mutate(retweet_to = str_extract(text, "(?<=@).+?(?=\\W)"))


retweet_df %>%
    group_by(retweet_to) %>%
    count() %>%
    arrange(desc(n)) %>%
    head(20) %>%
    ggplot(data = ., aes(x = retweet_to, y = n))+
    geom_bar(stat = "identity")+
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

#### Sentiments in Tweets over time

```r
source("text_to_sentiment.R")

sentiment_plot <- df %>%
  mutate(date = round_date(createdAt, "minute")) %>%
  group_by(date) %>%
  summarise(sentiment = text_to_sentiment(text)) %>%
  ggplot(data = ., aes(x  = date, y = sentiment)) +
  geom_line()+
  theme_bw()
```
